```python
# 1. Read the melb_data.csv file into a DataFrame.

import pandas as pd

#  Read the CSV file
df = pd.read_csv('melb_data.csv')
print(df)
```

```python

# 2. Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# 3 . Handle missing values in the specified columns as follows:
#   Use the mode to fill missing values in the Car and CouncilArea columns.
```

```python
car_mode = df['Car'].mode()[0]
df['Car'] = df['Car'].fillna(car_mode)

council_area_mode = df['CouncilArea'].mode()[0]
df['CouncilArea'] = df['CouncilArea'].fillna(council_area_mode)

# Confirm no missing values
print("Missing values after filling:")
print(df[['Car', 'CouncilArea']].isnull().sum())
```

```python
# 3 . Handle missing values in the specified columns as follows:
#   Use the mean to fill missing values in the BuildingArea and median to fill missing values YearBuilt columns.
```

```python
df["BuildingArea"] = df["BuildingArea"].fillna(df["BuildingArea"].mean())
df["YearBuilt"] = df["YearBuilt"].fillna(df["YearBuilt"].median())

# Verify that missing values are filled
print(df[["BuildingArea", "YearBuilt"]].isnull().sum())
```

```python
#4. Re-check for missing values to ensure that all missing values have been addressed
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# 5. Create the following visualizations:
# A histogram of the Price column.


import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("melb_data.csv")

# Plot histogram of Price
plt.figure(figsize=(10,6))
plt.hist(df["Price"].dropna(), bins=30, color="skyblue", edgecolor="black")

plt.title("Histogram of Property Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```

```python
# 5. Create the following visualizations:

#  A scatter plot of Price versus Distance.



import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("melb_data.csv")

# Scatter plot of Price vs Distance
plt.figure(figsize=(10,6))
plt.scatter(df["Distance"], df["Price"], alpha=0.6, color="orange", edgecolors="black")

plt.title("Scatter Plot of Price vs. Distance")
plt.xlabel("Distance")
plt.ylabel("Price")

plt.show()
```
